```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Continuous Features Analysis
def univariate_continuous(df, column, title):
    print(f"Summary Statistics for {column}:\n{df[column].describe()}\n")
    plt.figure(figsize=(10, 6))
    sns.histplot(df[column], kde=True, bins=30, color='skyblue')
    plt.title(f'Distribution of {title}', fontsize=14)
    plt.xlabel(column, fontsize=12)
    plt.ylabel('Frequency', fontsize=12)
    plt.grid(True)
    plt.show()

# Categorical Features Analysis
def univariate_categorical(df, column, title):
    print(f"Value Counts for {column}:\n{df[column].value_counts(normalize=True) * 100}\n")
    plt.figure(figsize=(8, 6))
    sns.countplot(data=df, x=column, palette='viridis')
    plt.title(f'Distribution of {title}', fontsize=14)
    plt.xlabel(column, fontsize=12)
    plt.ylabel('Count', fontsize=12)
    plt.grid(axis='y')
    plt.show()

# Univariate Analysis
print("Univariate Analysis")
univariate_continuous(fraud_data, "purchase_value", "Purchase Value")
univariate_categorical(fraud_data, "source", "Traffic Source")
univariate_categorical(fraud_data, "class", "Fraudulent Transactions")
```

```python
import pandas as pd

# Convert IP addresses to integer for fraud_data
def ip_to_int(ip):
    if isinstance(ip, str):
        try:
            # Convert IP address to integer (assuming it's in decimal format)
            ip_parts = ip.split('.')
            return sum(int(part) << (8 * (3 - idx)) for idx, part in enumerate(ip_parts))
        except ValueError:
            return None  # Return None if there's an issue converting the IP
    return None  # Return None if IP is not a string

# Convert IP addresses in fraud_data to integers, ensuring invalid IPs are handled
fraud_data['ip_address_int'] = fraud_data['ip_address'].apply(ip_to_int)

# Check the updated fraud_data with ip_address_int
print(fraud_data[['ip_address', 'ip_address_int']].head())

# Ensure that IP address ranges are correctly converted to integers in IP_address_data
IP_address_data['lower_bound_ip_address'] = pd.to_numeric(IP_address_data['lower_bound_ip_address'], errors='coerce')
IP_address_data['upper_bound_ip_address'] = pd.to_numeric(IP_address_data['upper_bound_ip_address'], errors='coerce')

# Verify the conversion of IP ranges to integers
print("Verify conversion of IP address ranges in IP_address_data")
print(IP_address_data[['lower_bound_ip_address', 'upper_bound_ip_address']].head())

# Merge fraud_data with IP_address_data based on IP range using 'ip_address' directly
print("Merging fraud_data with IP_address_data based on IP range")

def get_country_from_ip(ip):
    if isinstance(ip, str):
        ip_int = ip_to_int(ip)
        if pd.isna(ip_int):  # Check if ip_int is NaN (invalid IP)
            return 'Unknown'

        # Find matching country by checking if IP falls within the range
        country = IP_address_data[(IP_address_data['lower_bound_ip_address'] <= ip_int) & 
                                  (IP_address_data['upper_bound_ip_address'] >= ip_int)]
        
        if not country.empty:
            return country['country'].values[0]
    return 'Unknown'  # Return 'Unknown' if no match is found or if IP is invalid

# Apply the IP-to-country mapping
fraud_data['country'] = fraud_data['ip_address'].apply(get_country_from_ip)

# Check the updated fraud_data with country info
print(fraud_data[['ip_address', 'country']].head())

# Save the merged data to a CSV file
fraud_data.to_csv('merged_fraud_data_with_country.csv', index=False)
print("Merged data saved as 'merged_fraud_data_with_country.csv'")

# Cross-tabulation for Country and Fraud Class
print("Cross-tabulation for Country and Fraud Class")
cross_tab_country = pd.crosstab(fraud_data['country'], fraud_data['class'], normalize='index') * 100
print(cross_tab_country)
```

```python
import pandas as pd

# Load your data (assuming the DataFrame is named fraud_data)
fraud_data = pd.read_csv('../data/processed/fraud_data.csv')

# Step 1: Convert purchase_time to datetime format
fraud_data['purchase_time'] = pd.to_datetime(fraud_data['purchase_time'])

# Step 2: Transaction Frequency - Count number of transactions per user
transaction_frequency = fraud_data.groupby('user_id').size().reset_index(name='transaction_frequency')

# Step 3: Transaction Velocity - Time difference between consecutive transactions for the same user
fraud_data['purchase_time_diff'] = fraud_data.groupby('user_id')['purchase_time'].diff()

# Convert the time difference to a numerical value (in seconds)
fraud_data['purchase_time_diff'] = fraud_data['purchase_time_diff'].dt.total_seconds()

# Fill NaN values for the first transaction of each user with a default value
fraud_data['purchase_time_diff'].fillna(0, inplace=True)

# Calculate transaction velocity (could be the average time between transactions, for example)
transaction_velocity = fraud_data.groupby('user_id')['purchase_time_diff'].mean().reset_index(name='transaction_velocity')

# Step 4: Time-Based Features - Extract hour of day and day of week from purchase_time
fraud_data['hour_of_day'] = fraud_data['purchase_time'].dt.hour
fraud_data['day_of_week'] = fraud_data['purchase_time'].dt.dayofweek

# Step 5: Merge the new features back into the original fraud_data
fraud_data = fraud_data.merge(transaction_frequency, on='user_id', how='left')
fraud_data = fraud_data.merge(transaction_velocity, on='user_id', how='left')

# Step 6: Save the updated fraud_data with new features
fraud_data.to_csv('fraud_data_with_features.csv', index=False)

# Check the updated data
print(fraud_data[['user_id', 'purchase_time', 'transaction_frequency', 'transaction_velocity', 'hour_of_day', 'day_of_week']].head())
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler, LabelEncoder
from sklearn.preprocessing import OneHotEncoder

# Assuming fraud_data is already loaded (from previous steps)

# Step 1: Normalize/Scale Numeric Features
# Select numeric columns for scaling
numeric_columns = ['purchase_value', 'age', 'ip_address_int', 'transaction_frequency', 'transaction_velocity', 'hour_of_day', 'day_of_week']

# Min-Max Scaling
min_max_scaler = MinMaxScaler()
fraud_data[numeric_columns] = min_max_scaler.fit_transform(fraud_data[numeric_columns])

# Standardization
standard_scaler = StandardScaler()
fraud_data[numeric_columns] = standard_scaler.fit_transform(fraud_data[numeric_columns])

# Step 2: Encode Categorical Features
# Label Encoding for 'sex', 'source' (if necessary)
label_encoder = LabelEncoder()
fraud_data['sex'] = label_encoder.fit_transform(fraud_data['sex'])
fraud_data['source'] = label_encoder.fit_transform(fraud_data['source'])

# One-Hot Encoding for 'browser' and other categorical features
fraud_data = pd.get_dummies(fraud_data, columns=['browser'], drop_first=True)

# Step 3: Check the results
print(fraud_data.head())

# Step 4: Save the preprocessed data to CSV
fraud_data.to_csv('fraud_data_preprocessed.csv', index=False)

```
