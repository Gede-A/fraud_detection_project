```python
#Define a function for data loading and structure to data frame
import pandas as pd
import os

def load_data(filepath:str)->pd.DataFrame:
    if not os.path.exists(filepath):
         raise FileNotFoundError(f"File not found at: {filepath}")
    else:
        return pd.read_csv(filepath)

    
```

```python
#Define the filepath variable and assign a value for each variable where the csv data files are loccated.
fraud_data_filepath = '../data/raw/Fraud_Data.csv'
credit_card_filepath = '../data/raw/creditcard.csv'
IP_address_filepath = '../data/external/IpAddress_to_Country.csv'
```

```python
#Load the frau data and visualize the top five raws, get general information and descriptions like mean standared devation, min, max and quartiles
fraud_data = load_data(fraud_data_filepath)
credit_card =load_data(credit_card_filepath)
IP_address_data =load_data(IP_address_filepath)
```

```python
def general_info(loaded_data):
    print(loaded_data.head())
    print(loaded_data.info())
    print(loaded_data.describe())

print("General informations and the top raws of the fraud data")
general_info(fraud_data)    
```

```python
print("General informations and the top raws of the credit card data")
general_info(credit_card) 
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Continuous Features Analysis
def univariate_continuous(df, column, title):
    print(f"Summary Statistics for {column}:\n{df[column].describe()}\n")
    plt.figure(figsize=(10, 6))
    sns.histplot(df[column], kde=True, bins=30, color='skyblue')
    plt.title(f'Distribution of {title}', fontsize=14)
    plt.xlabel(column, fontsize=12)
    plt.ylabel('Frequency', fontsize=12)
    plt.grid(True)
    plt.show()

# Categorical Features Analysis
def univariate_categorical(df, column, title):
    print(f"Value Counts for {column}:\n{df[column].value_counts(normalize=True) * 100}\n")
    plt.figure(figsize=(8, 6))
    sns.countplot(data=df, x=column, palette='viridis')
    plt.title(f'Distribution of {title}', fontsize=14)
    plt.xlabel(column, fontsize=12)
    plt.ylabel('Count', fontsize=12)
    plt.grid(axis='y')
    plt.show()

# Univariate Analysis
print("Univariate Analysis")
univariate_continuous(fraud_data, "purchase_value", "Purchase Value")
univariate_categorical(fraud_data, "source", "Traffic Source")
univariate_categorical(fraud_data, "class", "Fraudulent Transactions")
```

```python
# Load IpAddress_to_Country.csv
ip_data = IP_address_data

# Categorical Analysis
print('Categorical Analysis')
print(f"Top 10 Countries by Frequency:\n{ip_data['country'].value_counts().head(10)}\n")
plt.figure(figsize=(12, 6))
ip_data['country'].value_counts().head(10).plot(kind='bar', color='teal')
plt.title("Top 10 Countries by Frequency of IP Ranges", fontsize=14)
plt.xlabel("Country", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
# Scatterplot for Purchase Value vs. Purchase Time
print('Scatterplot for Purchase Value vs. Purchase Time')
sns.scatterplot(x='purchase_time', y='purchase_value', hue='class', data=fraud_data, alpha=0.7)
plt.title('Purchase Value vs. Purchase Time by Fraud Class')
plt.xlabel('Purchase Time')
plt.ylabel('Purchase Value')
plt.legend(title='Fraud Class', loc='upper right')
plt.show()
```

```python
# Scatterplot for Time vs. Amount
print('Scatterplot for Time vs. Amount')
sns.scatterplot(x='Time', y='Amount', hue='Class', data=credit_card, alpha=0.7)
plt.title('Transaction Time vs. Amount by Fraud Class')
plt.xlabel('Time (seconds)')
plt.ylabel('Transaction Amount')
plt.legend(title='Fraud Class', loc='upper right')
plt.show()
```

```python
# Correlation between Age and Purchase Value
print('Correlation between Age and Purchase Value')
from scipy.stats import pearsonr
correlation, p_value = pearsonr(fraud_data['age'], fraud_data['purchase_value'])
print(f"Correlation between Age and Purchase Value: {correlation:.2f}, p-value: {p_value:.4f}")
```

```python
# Correlation between V1 and Amount
print('Correlation between V1 and Amount')
correlation, p_value = pearsonr(credit_card['V1'], credit_card['Amount'])
print(f"Correlation between V1 and Amount: {correlation:.2f}, p-value: {p_value:.4f}")
```

```python
# Cross-tabulation for Source and Fraud Class
print('Cross-tabulation for Source and Fraud Class')
cross_tab = pd.crosstab(fraud_data['source'], fraud_data['class'], normalize='index') * 100
print(cross_tab)

# Heatmap for visualization
print('Heatmap for visualization')
sns.heatmap(cross_tab, annot=True, fmt=".1f", cmap="coolwarm")
plt.title('Fraud Distribution by Source')
plt.ylabel('Source')
plt.xlabel('Fraud Class')
plt.show()
```

```python
# Convert IP addresses to integer for merging
print('Convert IP addresses to integer for merging')
IP_address_data['lower_bound_ip_address'] = IP_address_data['lower_bound_ip_address'].astype(int)
IP_address_data['upper_bound_ip_address'] = IP_address_data['upper_bound_ip_address'].astype(int)
# fraud_data['IP_address_data'] = fraud_data['IP_address_data'].astype(int)

# Map IP to Country
print('Map IP to Country')
def map_ip_to_country(ip):
    country = IP_address_data[(IP_address_data['lower_bound_ip_address'] <= ip) & (IP_address_data['upper_bound_ip_address'] >= ip)]['country']
    return country.values[0] if not country.empty else 'Unknown'

fraud_data['country'] = fraud_data['ip_address'].apply(map_ip_to_country)

# Cross-tabulation for Country and Fraud Class
print('Cross-tabulation for Country and Fraud Class')
cross_tab_country = pd.crosstab(fraud_data['country'], fraud_data['class'], normalize='index') * 100
print(cross_tab_country)
```
